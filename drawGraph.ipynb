```python
%%R

library('igraph')

dat <- read.table('Data/VariousTopics/edge_list.txt')
g <- graph.data.frame(dat,directed=T)
#pdf("network.pdf", width = 200, height = 200)
#plot(g, layout=layout.fruchterman.reingold.grid, vertex.size=1, 
#     edge.arrow.size=0.1, vertex.label=NA)

ec <- leading.eigenvector.community(g, steps = -1, weights = E(g)$weight,
                              start = NULL, options = igraph.arpack.default,
                              callback = NULL, extra = NULL, env = parent.frame())


plot(ec, g, vertex.label=NA,vertex.size=1,edge.arrow.size=0.1)

#dev.off()

#layout=layout.fruchterman.reingold(g, niter=10000)
```
