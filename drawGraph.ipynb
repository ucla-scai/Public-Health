```python
%%R

not.installed <- function(pkg) !is.element(pkg, installed.packages()[,1]) 
if (not.installed("igraph")) install.packages("igraph")
```

```python
%%R

library('igraph')

dat <- read.table('Data/VariousTopics/0.4_edge_list.txt', comment.char = "")
g <- graph.data.frame(dat,directed=T)
#pdf("network.pdf", width = 200, height = 200)
#plot(g, layout=layout.fruchterman.reingold.grid, vertex.size=1, 
#     edge.arrow.size=0.1, vertex.label=NA)

ec <- leading.eigenvector.community(g, steps = -1, weights = E(g)$weight,
                              start = NULL, options = igraph.arpack.default,
                              callback = NULL, extra = NULL, env = parent.frame())


plot(ec, g, vertex.label=NA,vertex.size=1,edge.arrow.size=0.1)

#dev.off()

#layout=layout.fruchterman.reingold(g, niter=10000)
```

```python
%%R

table(ec[[2]])
```

```python
%%R

numTopics = length(table(ec[[2]]))
topicTable = table(ec[[2]])

community_mat <-  matrix(, nrow = numTopics, ncol = 50)

for(topic in 1:numTopics) {
    curr_vec <- c()
    for(idx in 1:length(membership(ec))) {
        if(membership(ec)[idx] == topic) {
            curr_vec <- c(curr_vec, names(membership(ec))[idx]);
        }
    }
    #if(topicTable[topic] > 5 && topicTable[topic] < 100)
    community_mat[topic,] <- curr_vec[1:50]
}

write.table(community_mat, file = "Data/VariousTopics/0.4_topics.txt", sep = " ")
```
